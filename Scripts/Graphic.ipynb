```python
import pandas as pd
import numpy as np
import scipy
import matplotlib.pyplot as plt
```

```python
PuroA = pd.read_csv('zetaAur-eclipse_B_2019_10_06_01_37_16.dat', delimiter = ' ', header = None)
PuroA.columns = ['L','I', 'C']
PuroA['L'] = PuroA['L']*10 #unidades en Amstrong

CombinadoA = pd.read_csv('zetaAur-eclipse_B_2019_10_25_01_32_02.dat', delimiter = ' ', header = None)
CombinadoA.columns = ['L','I', 'C']
CombinadoA['L'] = CombinadoA['L']*10 #unidades en Amstrong

PuroDu = pd.read_csv('zetaAur-eclipse_B_2019_11_15_23_52_17.dat', delimiter = ' ', header = None)
PuroDu.columns = ['L','I', 'C']
PuroDu['L'] = PuroDu['L']*10 #unidades en Amstrong

PuroDe = pd.read_csv('zetaAur-eclipse_B_2019_12_14_19_43_59.dat', delimiter = ' ', header = None)
PuroDe.columns = ['L','I', 'C']
PuroDe['L'] = PuroDe['L']*10 #unidades en Amstrong
```

```python
#Graficamos todo el ezspectro en la banda del azul
plt.figure(figsize = [8,3])
plt.plot(PuroA['L'],PuroA['I'],marker=' ',ls = '-',label = 'Puro Antes')
plt.ylabel("Flujo",fontsize=18)
plt.xlabel("$\lambda [A]$",fontsize=18)
plt.ylim(-0.02, 0.05)
plt.xlim((3800,4050))
plt.legend(fontsize = 8)
plt.grid()
plt.show()

plt.figure(figsize = [8,3])
plt.plot(CombinadoA['L'],CombinadoA['I'],marker=' ',ls = '-', color = 'r', label = 'Combinado Antes')
plt.ylabel("Flujo",fontsize=18)
plt.xlabel("$\lambda [A]$",fontsize=18)
plt.ylim(-0.02, 0.15)
plt.xlim((3800,4050))
plt.legend(fontsize = 8)
plt.grid()
plt.show()

plt.figure(figsize = [8,3])
plt.plot(PuroDu['L'],PuroDu['I'],marker=' ',ls = '-',color = 'purple', label = 'Puro eclipse total')
plt.ylabel("Flujo",fontsize=18)
plt.xlabel("$\lambda [A]$",fontsize=18)
plt.ylim(-0.02, 0.06)
plt.xlim((3800,4050))
plt.legend(fontsize = 8)
plt.grid()
plt.show()

plt.figure(figsize = [8,3])
plt.plot(PuroDe['L'],PuroDe['I'],marker=' ',ls = '-', color = 'orange', label = 'Puro despues')
plt.ylabel("Flujo",fontsize=18)
plt.xlabel("$\lambda [A]$",fontsize=18)
plt.ylim(-0.02, 0.03)
plt.xlim((3800,4050))
plt.legend(fontsize = 8)
plt.grid()
plt.show()
```
