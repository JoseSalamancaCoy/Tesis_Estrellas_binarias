```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Ellipse
import pandas as pd
from scipy.optimize import curve_fit
from scipy import integrate
import numpy as np
```

```python
Data = pd.read_csv("photometry.txt",delimiter=',', skiprows=0)
```

```python
#Data
```

```python
DatosB = Data[Data['Band'] == 'B']
DatosTB = Data[Data['Band'] == 'TB']
```

```python
JD = DatosB['JD']
Mag = DatosB['Magnitude']
error = DatosB['Uncertainty']
Banda = DatosB['Band']


JD2 = DatosTB['JD']
Mag2 = DatosTB['Magnitude']
error2 = DatosTB['Uncertainty']
Banda2 = DatosTB['Band']
```

```python
fig = plt.figure(figsize=(15, 7))
plt.xlabel('Día Juliano', fontsize = 18)                  
plt.ylabel('Magnitud', fontsize = 18) 
(_, caps, _) = plt.errorbar(JD, Mag,error,0,marker ='.', ls ='', capsize=2, label='datos página')
(_, caps, _) = plt.errorbar(JD2, Mag2,error2,0,marker ='.', ls ='', capsize=2, label='datos página')

for cap in caps:
    cap.set_markeredgewidth(1)
plt.grid()
plt.ylim((4.7,5.6))
#plt.xlim((2.458821e+06,2.458840e+06))
plt.gca().invert_yaxis()
plt.legend(fontsize = 13, loc=0)
plt.show()
```
